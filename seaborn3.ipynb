```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd 
import seaborn as sns

df = pd.read_csv('./data/iris_synthetic.csv')
df.head()
numeric = df.select_dtypes(include=[np.number]) # Select only numeric columns
sns.clustermap(numeric,cmap='viridis', standard_scale=1) # Cluster map of the iris dataset ,standard scale 1 means each feature is centered and scaled to unit variance, cmap means the color map used for the visualization
```

```python
plot1= sns.PairGrid(df) # Create a PairGrid for the iris dataset
plot1.map(plt.scatter)# Scatter plot for each pair of features
# plot1.map_diag(sns.histplot) # Histogram for the diagonal
```
