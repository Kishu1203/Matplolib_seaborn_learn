```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
df1= pd.read_csv('./data/tips_synthetic.csv')
df1.head(10)
# df1.info()
```

```python
sns.pairplot(df1,hue='sex',palette='husl') # Pairplot of tips dataset and hue does is that it colors the points by the 'sex' column
```

```python
sns.rugplot(df1['tip'],height=0.1) # Rugplot for tips column and height adjustment
```

```python
print(sns.axes_style().keys()) # prints all the available styles in seaborn
sns.set_style('darkgrid')#style to darkgrid
sns.set_context('notebook', font_scale=1.5)# setting the context to notebook and increasing the font size
sns.despine(left=True,) # removes the left spine
sns.kdeplot(df1['tip'],shade=False,label='Tips') # Kernel Density Estimate plot for tips)
```

```python
sns.barplot(x='day',y='total_bill',data=df1,estimator=np.mean) # Barplot for average total bill per day with standard deviation as error bars 
```

```python
sns.countplot(df1,x='sex') # Countplot
```

```python
sns.boxplot(x='day',y='total_bill',data=df1,hue='sex',) # Boxplot for total bill per day
```

```python
sns.violinplot(x='day',y='total_bill',data=df1,hue='sex',split=True) # Violin plot for total bill per day
```

```python
sns.stripplot(x='day',y='total_bill',data=df1,hue='sex',jitter=True,dodge=True) # Strip plot for total bill per day with jitter, jitter does is add random noise to the data points and dodge does is separate the points for each hue category
plt.legend(loc=1) # Move the legend to the upper right corner, 0 is the default location, 1 is the upper right corner, 2 is the lower right corner, 3 is the upper left corner, 4 is the lower left corner 
```

```python
#Stacking of plots 
sns.violinplot(df1,x='day',y='total_bill')
sns.swarmplot(df1,x='day',y='total_bill') # Swarm plot for total bill per day
```

```python
plot1 = sns.FacetGrid(df1, col='day', row='time')
plot1.map(plt.hist, 'total_bill',bins=10) # FacetGrid for total bill and tip per day and time
plot1.map(plt.scatter, 'total_bill', 'tip') # Scatter plot for total bill and tip per day and time
```

```python
sns.lmplot(x='total_bill', y='tip', data=df1,markers='o',scatter_kws={'alpha':0.5}) # Linear model plot for total bill and tip
```

```python
sns.lmplot(x='total_bill', y='tip', data=df1,markers='o',scatter_kws={'alpha':0.5},col='day', row='time') # Linear model plot for total bill and tip per day and time
```
